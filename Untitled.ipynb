The class Sinusoidal_Wave defines the object wave which has the attributes:<br>
amp- stores amplitude<br>
freq- stores frequency<br>
phase- stores frequency of the wave<br>
It is used for sinusoidal waves<br>

```python
class Sinusoidal_Wave:
    def __init__(self, amp, period, phase):
        self.amp=amp
        self.freq=(2*np.pi)/period
        self.phase=phase
```

The function plot_SWave is used to plot the sinusoidal wave provided to it. For now takes the following:<br>
Wave: Takes the Sinusoidal_Wave object<br>
w_type: Takes in if the wave to be plotted is Sin or Cos, or phase or amplitude of the given waves<br>
start_time: Takes in the starting time from where the wave has to be plotted. Default=0<br>
end_Time: Takes in the ending time till where the wave has to be plotted. Deafult=10<br><br>

Prints what is being plotted.<br>
Takes a time array with default values=256<br>
theta corresponds to the final array obtained after calculating using wave frequency, time and phase<br>

```python
def plot_SWave(wave, w_type, start_time=0, end_time=10):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta=wave.freq*time+wave.phase
    
    if(w_type=='SIN'):        
        print('Plotting Sin wave of amp={}, period={:.2f}, phase={}'.format(wave.amp,2*np.pi/wave.freq,wave.phase))
        y = wave.amp*np.sin(theta)
        plt.xlabel('Time->')
        plt.ylabel('Sin(x)->')
        plt.plot(time, y)
        plt.show()
    elif(w_type=='COS'):
        print('Plotting Cos wave of amp={}, period={:.2f}, phase={}'.format(wave.amp,2*np.pi/wave.freq,wave.phase))
        y = wave.amp*np.cos(theta)
        plt.xlabel('Time->')
        plt.ylabel('Cos(x)->')
        plt.plot(time, y)
        plt.show()      
    elif(w_type=='PHASE'):
        print('Plotting phase of wave with amp={}, period={:.2f}, phase={}'.format(wave.amp,2*np.pi/wave.freq,wave.phase))
        plt.xlabel('Time->')
        plt.ylabel('Phase->')
        theta=wave.freq*(time+((2*np.pi)/wave.freq)/2)+wave.phase
        plt.plot(time, -(np.pi-theta%(2*np.pi)))
        plt.show()
    elif(w_type=='AMP'):
        print('Plotting amplitude of wave with amp={}, period={:.2f}, phase={}'.format(wave.amp,2*np.pi/wave.freq,wave.phase))
        plt.xlabel('Time->')
        plt.ylabel('Amplitude->')
        plt.plot(time,wave.amp*np.ones(time.shape))
        plt.show()     
```

```python
SinWave1=Sinusoidal_Wave(1,2,0)
```

```python
plot_SWave(SinWave1,'PHASE',-2)
```

```python
def Add_SWave(wave1, wave2, w_type, start_time=0, end_time=10):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta1=wave1.freq*time+wave1.phase
    theta2=wave2.freq*time+wave2.phase
    
    if(w_type=='SIN'):        
        print('Plotting Sin wave of amp={}, period={:.2f}, phase={} added with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.sin(theta1)
        y2 = wave2.amp*np.sin(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Sin(x)->')
        plt.plot(time,y1)
        plt.plot(time,y2)
        plt.plot(time, y1+y2)
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()
    elif(w_type=='COS'):
        print('Plotting Cos wave of amp={}, period={:.2f}, phase={} added with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.cos(theta1)
        y2 = wave2.amp*np.cos(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Cos(x)->')
        plt.plot(time,y1)
        plt.plot(time,y2)
        plt.plot(time, y1+y2)
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()  
```

```python
SinWave1=Sinusoidal_Wave(1,2,0)
SinWave2=Sinusoidal_Wave(2,2,2*np.pi)
Add_SWave(SinWave1,SinWave2,'SIN')
```

```python
def Multi_SWave(wave1, wave2, w_type, start_time=0, end_time=10):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta1=wave1.freq*time+wave1.phase
    theta2=wave2.freq*time+wave2.phase
    
    if(w_type=='SIN'):        
        print('Plotting Sin wave of amp={}, period={:.2f}, phase={} multiplied with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.sin(theta1)
        y2 = wave2.amp*np.sin(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Sin(x)->')
        plt.plot(time,y1,'b-')
        plt.plot(time,y2,'y-')
        plt.plot(time, y1*y2,'r')
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()
    elif(w_type=='COS'):
        print('Plotting Cos wave of amp={}, period={:.2f}, phase={} multiplied with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.cos(theta1)
        y2 = wave2.amp*np.cos(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Cos(x)->')
        plt.plot(time,y1)
        plt.plot(time,y2)
        plt.plot(time, y1*y2)
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()  
```

```python
SinWave1=Sinusoidal_Wave(1,2,0)
SinWave2=Sinusoidal_Wave(2,2,2*np.pi)
Multi_SWave(SinWave1,SinWave2,'SIN')
```

```python
def Sub_SWave(wave1, wave2, w_type, start_time=0, end_time=10):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta1=wave1.freq*time+wave1.phase
    theta2=wave2.freq*time+wave2.phase
    
    if(w_type=='SIN'):        
        print('Plotting Sin wave of amp={}, period={:.2f}, phase={} subtracted with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.sin(theta1)
        y2 = wave2.amp*np.sin(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Sin(x)->')
        plt.plot(time,y1)
        plt.plot(time,y2)
        plt.plot(time, y1-y2)
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()
    elif(w_type=='COS'):
        print('Plotting Cos wave of amp={}, period={:.2f}, phase={} subtracted with wave amp={}, period={:.2f}, phase={}'.format(wave1.amp,2*np.pi/wave1.freq,wave1.phase, 
                                                                                                                            wave2.amp,2*np.pi/wave2.freq,wave2.phase))
        y1 = wave1.amp*np.cos(theta1)
        y2 = wave2.amp*np.cos(theta2)
        plt.xlabel('Time->')
        plt.ylabel('Cos(x)->')
        plt.plot(time,y1)
        plt.plot(time,y2)
        plt.plot(time, y1-y2)
        plt.legend(["wave1","wave2","resultant"],loc ="lower right")
        plt.show()  
```

```python
SinWave1=Sinusoidal_Wave(1,2,0)
SinWave2=Sinusoidal_Wave(2,2,2*np.pi)
Sub_SWave(SinWave1,SinWave2,'SIN')
```

```python
def plot_rectpulse(amp,start_time,end_time,time_period,width):
    x=np.linspace(start_time,end_time,256,endpoint=True)
    width=width/2
    y=[1 if ((num+width)%time_period<2*width and (num+width)%time_period>-2*width) else 0 for num in x]
    plt.xlabel('Time->')
    plt.ylabel('x(t)->')
    plt.plot(x,y)
    plt.show()
```

```python
plot_rectpulse(1,-3,3,2,1)
```

```python
def plot_sinc(start_time,end_time):
    x=np.linspace(start_time,end_time,256,endpoint=True)
    y=[(np.sin(np.pi*x_))/(np.pi*x_) if x_!=0 else 1 for x_ in x]
    plt.xlabel('Time->')
    plt.ylabel('x(t)->')
    plt.plot(x,y)
    plt.show()
```

```python
plot_sinc(-3*np.pi,3*np.pi)
```

```python
def plot_delta(wave,start_time=0, end_time=10):   
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta=wave.freq*time+wave.phase
    print('Plotting Cos wave of amp={}, period={:.2f}, phase={}'.format(wave.amp,2*np.pi/wave.freq,wave.phase))
    #y = wave.amp*np.cos(theta)
    y=0
    for x in range(0,3):
        y+=wave.amp*np.cos(x*theta)
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y)
    plt.show()
    
    
    y=0
    for x in range(0,6):
        y+=wave.amp*np.cos(x*theta)
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y)
    plt.show()
    
    y=0
    for x in range(0,10):
        y+=wave.amp*np.cos(x*theta)
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y)
    plt.show()
    
    y=0
    for x in range(0,100):
        y+=wave.amp*np.cos(x*theta)
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y)
    plt.show()
```

```python
def am_sc(wave1,wave2,start_time,end_time):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta1=wave1.freq*time+wave1.phase
    theta2=wave2.freq*time+wave2.phase
    y1 = wave1.amp*np.cos(theta1)
    y2 = wave2.amp*np.cos(theta2)
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y1*y2)
    plt.show()
```

```python
am_sc(message,carrier,0,3)
```

```python
def am_wc(wave1,wave2,start_time,end_time):
    time = np.linspace(start_time, end_time, 256, endpoint=True)
    theta1=wave1.freq*time+wave1.phase
    theta2=wave2.freq*time+wave2.phase
    y1 = wave2.amp+wave1.amp*np.cos(theta1)
    y2 = 1*np.cos(theta2) 
    plt.xlabel('Time->')
    plt.ylabel('Cos(x)->')
    plt.plot(time, y1*y2)
    plt.show()
```

```python
modulator_frequency = 4.0
carrier_frequency = 40.0
modulation_index = 1.0

time = np.arange(44100.0) / 44100.0
modulator = np.sin(2.0 * np.pi * modulator_frequency * time) * modulation_index
carrier = np.sin(2.0 * np.pi * carrier_frequency * time)
product = np.zeros_like(modulator)

for i, t in enumerate(time):
    product[i] = np.sin(2. * np.pi * (carrier_frequency * t + modulator[i]))


plt.subplot(3, 1, 1)
plt.plot(modulator)
plt.subplot(3, 1, 2)
plt.plot(carrier)
plt.subplot(3, 1, 3)
plt.plot(product)
plt.show()
```
